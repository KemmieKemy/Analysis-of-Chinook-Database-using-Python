# DATA ANALYSIS OF THE CHINOOK DATABASE USING PYTHON

# The city that has the best customers;
This program returns 1 city that has the highest sum of invoice totals.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
cu = pd.read_csv('Customer.csv')
inv = pd.read_csv('Invoice.csv')
invL = pd.read_csv('InvoiceLine.csv')
tr = pd.read_csv('Track.csv')
gen = pd.read_csv('Genre.csv')
inv = inv.set_index ('InvoiceId')
#Which city has the best customers?
print ('The City with the Best Customers is:')
inv.groupby('BillingCity')['Total'].sum() .sort_values(ascending = False).head(1)
```

```python
cu_plot = inv.groupby(['BillingCity'])['Total'].sum().sort_values(ascending = False).head(10)
plt.xticks(color='blue')
plt.yticks(color='blue')
plt.xlabel('Cities', color='red')
plt.ylabel('Total Invoice', color='red')
plt.title('10 Cities that have the best customers', color='blue')
cu_plot.plot.bar(color='red', edgecolor= 'blue')
plt.show()
```

# The countries that have the most invoices
This program provides a table of *BillingCountry* and *Invoices* ordered by the number of invoices for each country. 

```python
#Which Countries have the most Invoices?
print ('The Countries with the most Invoices are:')
inv.groupby('BillingCountry')['BillingCountry'].count().sort_values(ascending = False)
```

```python
country_plot = inv.groupby('BillingCountry')['BillingCountry'].count().sort_values(ascending = False).head(10)
plt.xticks(color='green')
plt.yticks(color='green')
plt.xlabel('Country', color='red')
plt.ylabel('Total Invoice', color='red')
plt.title('10 Countries with the most Invoices', color='red')
country_plot.plot.bar(color='yellow', edgecolor='black')
plt.show()
```

# The best customer
This program returns the customer who has spent the most money.


```python
#Who is the Best Customer?
cu_in = cu.merge(inv, on='CustomerId')
cu_in.groupby(['FirstName', 'LastName'])['Total'].sum() .sort_values(ascending = False).head(1)
```

```python
plt.style.use('ggplot')
cu_in_plot = cu_in.groupby(['FirstName', 'LastName'])['Total'].sum() .sort_values(ascending = False).head(10)
plt.xticks(color='green')
plt.yticks(color='green')
plt.xlabel('Name', color='red')
plt.ylabel('Total Invoice', color='red')
plt.title('Graph of Total Invoice against Customer', color='red')
cu_in_plot.plot.bar(color='yellow')
plt.show()
```

# Details of all Rock Music listeners;
This program will output the email, first name, last name, and Genre of all Rock Music listeners ordered alphabetically by email address. 

```python
cu.join(inv, on=['CustomerId'], rsuffix ='cu', how='inner')
inv.join(invL, on=['InvoiceId'], rsuffix ='inv', how='inner')
invL.join(tr, on=['TrackId'], rsuffix ='invL', how='inner')
tr.join(gen, on=['GenreId'], rsuffix ='tr', how='inner')
#Rock Music Listeners
print ('List of Rock Music Listeners:')
cu.sort_values('Email', inplace=True)
#filtering the Customer dataframe
cu_filter = gen['Name'] == 'Rock'
cu.where(cu_filter, inplace=False)
print (cu[['Email','FirstName','LastName']])
```
